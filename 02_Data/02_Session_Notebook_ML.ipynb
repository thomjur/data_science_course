## II. Opening a Comma-Separated Value (CSV) File

```python
### version 1: Python built-in csv reader
import csv

with open('data_sources/fruits.csv') as f:
    csv_file = csv.reader(f, delimiter=",")
    for row in csv_file:
        print(row)
```
